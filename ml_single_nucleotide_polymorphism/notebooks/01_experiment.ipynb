```python
# Technique 1: Simple Forward Fill / Backward Fill
def forward_fill_nan(df):
    """Fill NaN with previous value in column"""
    return df.fillna(method='ffill').fillna(method='bfill')

# Technique 2: Mean/Median/Mode Imputation
def simple_impute(df, strategy='median'):
    """Simple imputation using mean, median, or mode"""
    from sklearn.impute import SimpleImputer
    
    # Separate numeric and categorical
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    categorical_cols = df.select_dtypes(include=['object', 'bool']).columns
    
    imputed_df = df.copy()
    
    # Impute numeric with mean/median
    if len(numeric_cols) > 0:
        numeric_imputer = SimpleImputer(strategy=strategy)
        imputed_df[numeric_cols] = numeric_imputer.fit_transform(df[numeric_cols])
    
    # Impute categorical with mode
    if len(categorical_cols) > 0:
        categorical_imputer = SimpleImputer(strategy='most_frequent')
        imputed_df[categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])
    
    return imputed_df

# Technique 3: KNN Imputation
def knn_impute(df, n_neighbors=5):
    """K-Nearest Neighbors imputation"""
    from sklearn.impute import KNNImputer
    
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    imputed_df = df.copy()
    
    if len(numeric_cols) > 0:
        knn_imputer = KNNImputer(n_neighbors=n_neighbors)
        imputed_df[numeric_cols] = knn_imputer.fit_transform(df[numeric_cols])
    
    return imputed_df

# Technique 4: Iterative Imputation (MICE)
def iterative_impute(df):
    """Iterative Imputation using MICE"""
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer
    
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    imputed_df = df.copy()
    
    if len(numeric_cols) > 0:
        iterative_imputer = IterativeImputer(random_state=42, max_iter=10)
        imputed_df[numeric_cols] = iterative_imputer.fit_transform(df[numeric_cols])
    
    return imputed_df

# Technique 5: Drop rows/columns with NaN
def drop_nan(df, axis=0, thresh=None):
    """Drop rows (axis=0) or columns (axis=1) with NaN"""
    if thresh:
        return df.dropna(axis=axis, thresh=thresh)
    return df.dropna(axis=axis)

print("Imputation functions defined!")
```

## Why Test Data Has NaN But Training Didn't

**The Key Difference:**

### During Training (Phase 1-2):
1. ✅ Data is loaded fresh - all values (0, 1, 2) are valid
2. ✅ Features are identified as binary or multi-category based on unique values
3. ✅ Multi-category features (with 3+ values) get **one-hot encoded** automatically
4. ✅ Binary features (with 2 values) stay as-is
5. ✅ Missing values are imputed normally

### During Replication/Test (Phase 8):
1. ⚠️ STREAMLINE loads the **binary_categorical_dict** from training
2. ⚠️ It expects binary features to have **EXACTLY** the same values as training
3. ⚠️ If a binary feature has a NEW value (like `2` when training only had `0` and `1`), it gets **replaced with NaN**
4. ⚠️ This happens in `model_replicate.py` lines 164-173
5. ⚠️ These NaN values are created AFTER data load but BEFORE imputation
6. ❌ **The imputation should handle them, but some slip through!**

**The Root Cause:**
- Your original data has features with values `{0, 1, 2}` 
- During training, these were identified as "binary" (probably because they had mostly 0/1 with occasional 2)
- During test, when value `2` appears, STREAMLINE sees it as "unexpected" and replaces with NaN
- The imputation code has fallback logic (lines 576-595) but it's not catching all cases

**Solution:** Ensure imputation runs properly on ALL NaN values before model evaluation.

### The Fix: Ensure Complete Imputation

The issue is that NaN values created during replication need to be properly imputed. The code has fallback logic, but we need to ensure it runs correctly. Here are the solutions:

```python
# SOLUTION: Fix the processed file by ensuring all NaN are imputed
# This mimics what STREAMLINE should do but ensures it's done correctly

def fix_nan_in_processed_file(df_processed, cat_features=None, quant_features=None):
    """
    Comprehensive NaN fixing function that handles all cases
    """
    df_fixed = df_processed.copy()
    
    # Identify feature types if not provided
    if cat_features is None or quant_features is None:
        # Boolean columns are categorical
        bool_cols = df_fixed.select_dtypes(include=['bool']).columns.tolist()
        # Integer columns that are not boolean
        int_cols = df_fixed.select_dtypes(include=['int64']).columns.tolist()
        # Float columns (these might have NaN)
        float_cols = df_fixed.select_dtypes(include=['float64']).columns.tolist()
        
        # Remove non-feature columns
        non_feature_cols = ['sample_id', 'label', 'Class']
        bool_cols = [c for c in bool_cols if c not in non_feature_cols]
        int_cols = [c for c in int_cols if c not in non_feature_cols]
        float_cols = [c for c in float_cols if c not in non_feature_cols]
        
        cat_features = bool_cols + [c for c in int_cols if df_fixed[c].nunique() <= 10]  # Heuristic
        quant_features = float_cols + [c for c in int_cols if c not in cat_features]
    
    print(f"Categorical features: {len(cat_features)}")
    print(f"Quantitative features: {len(quant_features)}")
    
    # Step 1: Impute categorical with mode
    print("\nImputing categorical features...")
    for col in cat_features:
        if col in df_fixed.columns and df_fixed[col].isna().sum() > 0:
            mode_val = df_fixed[col].mode()
            fill_val = mode_val[0] if len(mode_val) > 0 else False  # False for bool, 0 for int
            if pd.isna(fill_val):
                fill_val = False if df_fixed[col].dtype == 'bool' else 0
            df_fixed[col].fillna(fill_val, inplace=True)
    
    # Step 2: Impute quantitative with median
    print("Imputing quantitative features...")
    for col in quant_features:
        if col in df_fixed.columns and df_fixed[col].isna().sum() > 0:
            median_val = df_fixed[col].median()
            fill_val = median_val if not pd.isna(median_val) else 0
            df_fixed[col].fillna(fill_val, inplace=True)
    
    # Step 3: Final pass - handle any remaining NaN
    print("Final pass - handling any remaining NaN...")
    remaining_nan_cols = df_fixed.columns[df_fixed.isna().any()].tolist()
    for col in remaining_nan_cols:
        if col not in ['sample_id', 'label', 'Class']:
            if df_fixed[col].dtype == 'bool':
                df_fixed[col].fillna(False, inplace=True)
            elif df_fixed[col].dtype in ['int64', 'int32']:
                df_fixed[col].fillna(0, inplace=True)
            else:
                mode_val = df_fixed[col].mode()
                fill_val = mode_val[0] if len(mode_val) > 0 else 0
                df_fixed[col].fillna(fill_val, inplace=True)
    
    return df_fixed

# Apply the fix
print("=== Applying Comprehensive NaN Fix ===")
df_fixed = fix_nan_in_processed_file(df_processed)

print(f"\n✅ NaN count after fix: {df_fixed.isna().sum().sum()}")
print(f"✅ All NaN values handled!")

# Verify no NaN remain
if df_fixed.isna().sum().sum() == 0:
    print("\n🎉 SUCCESS: All NaN values have been imputed!")
else:
    print(f"\n⚠️ WARNING: {df_fixed.isna().sum().sum()} NaN values still remain")
    print("Remaining NaN columns:", df_fixed.columns[df_fixed.isna().any()].tolist())
```
